```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
time_labels = np.load('time_labels_50hz.npy')
time_labels[10]
```

```python
data = np.load('Xhat_centeredmean_unshifted.npy')
data = np.mean(data, axis=2)
data = data[:,:,10:20]
data = np.mean(data, axis=2)
```

```python
x_labels = [0, 22.5, 45, 67.5, 90, 112.5, 135, 157.5]

custom_colors = ['blue', 'green', 'red', 'purple', 'orange', 'yellow', 'brown', 'pink']

sns.set_palette(custom_colors)

fig, ax = plt.subplots(figsize=(8, 6))

# Plot each line and assign labels
for i in range(8):
    color = custom_colors[i]
    plt.plot(data[:, i], label=f'{x_labels[i]}°', color = color)
    plt.axvline(x=i, linestyle='--', linewidth=0.5, color = color)

# Add labels to x and y axes
ax.set_xticks(range(len(x_labels)))
ax.set_xticklabels(x_labels)
ax.set_xlabel('Orientation (degrees)')
ax.set_ylabel('Channel Response (a. u.)')

# Add a title to the plot
ax.set_title('Unshifted orientation channels')

# Place the legend at the right side
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.show()
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
sorteddesign = np.load('sortedesign.npy', allow_pickle=True)
sorteddesign = sorteddesign.T

weights = np.load('weights.npy', allow_pickle=True)
weights = weights.T

time_labels = np.load('time_labels_50Hz.npy', allow_pickle=True)
time_labels
```

```python
orientation = [180/8*i for i in range(8)]
orientation
```

```python
plt.imshow(sorteddesign, aspect = 'auto', origin='lower', cmap='viridis')
plt.xticks(np.arange(len(orientation)), orientation)
plt.xlabel('Orientation (degrees)')
plt.yticks([])

plt.ylabel('Sorted Design')
plt.title('Sorted Design Matrix')


plt.colorbar(label='Channel response (a. u.)')
```

```python
numC = 8
angles = [i * 180./numC for i in range(numC)]

x_labels = np.array(angles)
```

```python
fig, ax = plt.subplots(figsize=(8, 6)) 
im = ax.imshow(weights, origin='lower')
ax.set_yticks(range(len(time_labels2)), time_labels2)
ax.set_yticks(ax.get_yticks()[::8])
ax.set_ylabel('Time (ms)')
ax.set_xlabel('Orientation (degrees)')  
x_labels = x_labels.astype(int)
ax.set_xticks(range(len(x_labels)), x_labels)
ax.set_xticks(ax.get_xticks()[::2])

cbar = fig.colorbar(im, ax=ax)
cbar.set_label('Weight (a. u.)')
ax.set_title('Model weights')

# plt.tight_layout()
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt

x = np.linspace(-2*np.pi, np.pi , 100)

y1 = np.sin(x)*2
y2 = np.sin(x)*1.5
y3 = np.sin(x)

plt.plot(x[0:34], y1[0:34], label='cardinal')
plt.plot(x[66:100], y3[66:100], label='diagonal')
plt.plot(x[33:67], y2[33:67], color='k', linestyle='--')

plt.axhline(y=0, color='k')
plt.legend()

plt.gca().set_yticklabels([])
plt.gca().set_xticklabels([])
# remove the ticks
plt.tick_params(top=False, bottom=False, left=False, right=False, labelleft=False, labelbottom=False)

plt.show()

```

```python
import scipy.stats as stats

mu1 = -2
sigma1 = 0.5
mu2 = 2
sigma2 = 0.5

x = np.linspace(-5, 5, 100)

y1 = stats.norm.pdf(x, mu1, sigma1)
y2 = stats.norm.pdf(x, mu2, sigma2)*0.5

plt.plot(x, y1, label = 'cardinal')
plt.plot(x, y2, label = 'diagonal')
plt.legend()
plt.show()
```
